```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from darts import TimeSeries
from darts.utils.statistics import check_seasonality,extract_trend_and_seasonality,plot_acf,plot_hist
from orbit.eda import eda_plot
```

```python
df = pd.read_csv('/content/airline-passengers.csv')
```

```python
df.shape
```

```python
series = TimeSeries.from_dataframe(df,'Month','Passengers')
```

```python
series.plot()
```

```python
check_seasonality(series)
```

```python
extract_trend_and_seasonality(series)[0].plot()
```

```python
extract_trend_and_seasonality(series)[1].plot()
```

```python
plot_acf(series)
```

```python
plot_hist(series)
```

```python
_ = eda_plot.ts_heatmap(df=df, date_col='Month', value_col='Passengers')
plt.show()
```

```python
_ = eda_plot.correlation_heatmap(df, var_list=['Passengers'], fig_width=10, fig_height=8)
plt.show()
```
